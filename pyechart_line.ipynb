```python
import pandas as pd
import numpy as np
from pyecharts.charts import Line
from pyecharts import options as opts
from pyecharts.render import make_snapshot
df = pd.read_excel('D:\PycharmProjects\Portfolio\AllGender.xlsx',index_col='Geography',)
select = [2021]
tag = df[['Year']].applymap(lambda x:x in select).any(1)
f = df[tag]
print(f)
select1 = [2016]
tag1 = df[['Year']].applymap(lambda x:x in select1).any(1)
f1 = df[tag1]
print(f1)
select2 = [2011]
tag2 = df[['Year']].applymap(lambda x:x in select2).any(1)
f2 = df[tag2]
print(f2)
select3 = [2006]
tag3 = df[['Year']].applymap(lambda x:x in select3).any(1)
f3 = df[tag3]
print(f3)
line = Line()
```

```python
a = df.head(8)
line = Line()
b = f.iloc[:,3:4]
c = np.array(b)
d = c.tolist()

b1 = f1.iloc[:,3:4]
c1 = np.array(b1)
d1 = c1.tolist()

b2 = f2.iloc[:,3:4]
c2 = np.array(b2)
d2 = c2.tolist()

b3 = f3.iloc[:,3:4]
c3 = np.array(b3)
d3 = c3.tolist()
```

```python

line.add_xaxis(b.index.to_list())
line.add_yaxis('2021 No certificate, diploma or degree',y_axis=d)
line.add_yaxis('2016 No certificate, diploma or degree',y_axis=d1)
line.add_yaxis('2011 No certificate, diploma or degree',y_axis=d2)
line.add_yaxis('2006 No certificate, diploma or degree',y_axis=d3)
line.set_global_opts(
    title_opts=opts.TitleOpts(title='Percentage'),
    toolbox_opts=opts.TooltipOpts(trigger='axis',axis_pointer_type='cross'),
    yaxis_opts=opts.AxisOpts(min_='dataMin')
    
)
line.render_notebook()
```

```python
make_snapshot(line.render(),file_name='pectest.gif',output_name='pectest.gif')
```
